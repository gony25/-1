```
from pyspark.ml.feature import VectorAssembler
```

```
train_cols = df.columns[:-1]
```

```
train_cols
```

```
vector = VectorAssembler(inputCols=train_cols, outputCol='features')
```

```
df = vector.transform(df)
```

```
df.show()
```

```
df = df.select(['features', 'SalePrice'])
```

```
df.show()
```

```
(train_df, test_df) = df.randomSplit([0.8, 0.2])
```

```
import pandas as pd
```

```
from pyspark.ml.regression import LinearRegression
```

```
lr = LinearRegression(featuresCol='features', labelCol='SalePrice')
```

```
lr_model = lr.fit(train_df)
```

```
trainingSummary =lr_model.summary
```

```
trainingSummary.meanAbsoluteError
```

```
trainingSummary.meanSquaredError
```

```
trainingSummary.rootMeanSquaredError
```

```
lr_predictions = lr_model.transform(test_df)
```

```
lr_predictions.select(['prediction', 'SalePrice']).show()
```

```
from pyspark.ml.evaluation import RegressionEvaluator
```

```
lr_evaluator = RegressionEvaluator(predictionCol = 'prediction', labelCol='SalePrice', metricName = 'mse')
```

```
lr_evaluator.evaluate(lr_predictions)
```
